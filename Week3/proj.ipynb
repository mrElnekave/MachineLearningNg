```python

def sigmoid(z):
    """
    Compute the sigmoid of z

    Args:
        z (ndarray): A scalar, numpy array of any size.

    Returns:
        g (ndarray): sigmoid(z), with the same shape as z
         
    """
          
    ### START CODE HERE ### 
    
    if type(z) != np.ndarray:
        z = np.array([z])
    g = np.zeros(z.shape[0])
    g = 1 / (1 + np.exp(-z))
    
    ### END SOLUTION ###  
    
    return g

```

```python
def compute_cost(X, y, w, b, *argv):
    """
    Computes the cost over all examples
    Args:
      X : (ndarray Shape (m,n)) data, m examples by n features
      y : (ndarray Shape (m,))  target value 
      w : (ndarray Shape (n,))  values of parameters of the model      
      b : (scalar)              value of bias parameter of the model
      *argv : unused, for compatibility with regularized version below
    Returns:
      total_cost : (scalar) cost 
    """

    m, n = X.shape
    
    ### START CODE HERE ###
    
    total_cost = 0
    
#     hypotheses = np.zeros(m)
#     for i in range(m):
#         hypotheses[i] = sigmoid(np.dot(X[i], w) + b)
    hypotheses = sigmoid(np.dot(X, w) + b)
    
    hypoth_1minus = np.log(1 - hypotheses) 
    hypotheses = np.log(hypotheses)
    
    losses = np.dot(y, hypotheses) + np.dot((1-y), hypoth_1minus)
    
    total_cost = -1/m * np.sum(losses)
    ### END CODE HERE ### 

    return total_cost
```

```python
def compute_gradient(X, y, w, b, *argv): 
    """
    Computes the gradient for logistic regression 
 
    Args:
      X : (ndarray Shape (m,n)) data, m examples by n features
      y : (ndarray Shape (m,))  target value 
      w : (ndarray Shape (n,))  values of parameters of the model      
      b : (scalar)              value of bias parameter of the model
      *argv : unused, for compatibility with regularized version below
    Returns
      dj_dw : (ndarray Shape (n,)) The gradient of the cost w.r.t. the parameters w. 
      dj_db : (scalar)             The gradient of the cost w.r.t. the parameter b. 
    """
    m, n = X.shape
    dj_dw = np.zeros(w.shape)
    dj_db = 0.

    ### START CODE HERE ### 
    for i in range(m):
#         z_wb = 0.
#         for j in range(n): 
#             z_wb += X[i][j] * w[j]
#         z_wb += b
#         f_wb = sigmoid(z_wb)
        f_wb = sigmoid(np.dot(X[i], w) + b)

        dj_db_i = f_wb - y[i]
        dj_db += dj_db_i
        
        for j in range(n):
            dj_dw[j] += dj_db_i * X[i][j]
            
    dj_dw = 1/m * dj_dw
    dj_db = 1/m * dj_db
    ### END CODE HERE ###

        
    return dj_db, dj_dw
```

```python
def predict(X, w, b): 
    """
    Predict whether the label is 0 or 1 using learned logistic
    regression parameters w
    
    Args:
      X : (ndarray Shape (m,n)) data, m examples by n features
      w : (ndarray Shape (n,))  values of parameters of the model      
      b : (scalar)              value of bias parameter of the model

    Returns:
      p : (ndarray (m,)) The predictions for X using a threshold at 0.5
    """
    # number of training examples
    m, n = X.shape   
    p = np.zeros(m)
   
    ### START CODE HERE ### 
    # Loop over each example
    for i in range(m):   
        z_wb = np.dot(X[i], w) + b

        # Apply the threshold
        p[i] = 0 if z_wb < 0 else 1
    ### END CODE HERE ### 
    return p
```
